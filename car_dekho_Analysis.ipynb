# Calculating depreciation for each vehicle and finding the most and least depreciated vehicles

```python
# 12) Most and least cost depreciated vehicle
df['Depreciation'] = df['Present_Price'] - df['Selling_Price']
df
```

```python
# The most depreciated vehicle
most_depreciated_vehicle = df.loc[df['Depreciation'].idxmax()]['Car_Name']
most_depreciated_vehicle
```

```python
# The location and row of the highest depreciated vehicle
df.loc[df.Depreciation == df.Depreciation.max()]
```

```python
# The least depreciated vehicle
least_depreciated_vehicle = df.loc[df['Depreciation'].idxmin()]['Car_Name']
least_depreciated_vehicle
```

```python
# The location and the row of the lowest depreciated vehicle
df.loc[df.Depreciation == df.Depreciation.min()]
```

# Finding brands with the least average depreciation

```python
# 13) Brands less affected by cost depreciation
brands_less_affected = df.groupby('Car_Name')['Depreciation'].mean().sort_values().head()
brands_less_affected
```

# Generating a correlation matrix to identify factors affecting cost depreciation

```python
# 14) Factors affecting cost depreciation
numeric_df = df.select_dtypes(include=['number'])
corr = numeric_df.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Filtering the dataset for two-wheelers and finding the oldest bike sold

# Filtering the two wheelers dataframe

```python
# Reseting the two wheelers data
two_wheelers.reset_index(drop = True, inplace = True)
two_wheelers.head(10)
```

```python
# Filtering cars from the two wheelers data
two_wheelers = two_wheelers.loc[two_wheelers.Car_Name != "alto k10"]
two_wheelers = two_wheelers.loc[two_wheelers.Car_Name != "800"]
two_wheelers = two_wheelers.loc[two_wheelers.Car_Name != "omni"]
two_wheelers.head(10)
```
